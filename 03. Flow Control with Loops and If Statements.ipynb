# If Statments

An if statement allows to perform flow control, creating a condition under which a specific action/task is only executed if a condition (test expression) is true.

```python
YourSalary = 2000
ProposedSalary = 2200
```

```python
if ProposedSalary > YourSalary:
    print("Cool, I'll think about it")
```

```python
AvgSalary = 2300
```

```python
if ProposedSalary >= AvgSalary:
    print('Great !')
else:
    print("Ehhh, I'll need to negotiate that...")
```

# Loops While

The While executes repeatedly an action as long as a condition remains true.

```python
TSLstockprice = 825

while TSLstockprice < 850:
    TSLstockprice += 2
    print('Hold', TSLstockprice)
    if TSLstockprice >= 850:
        print('Sell!!')
```

# Loops For

The for iterates over elements in a sequence (list, tuples, dict, etc)

```python
stockCompanies = ['Facebook', 'Apple', 'Amazon', 'Netflix', 'Google']
priceChange = [1.1, -2.2, -0.9, 4.2, 2.7]
```

```python
for company in stockCompanies:
    print(company)
```

```python
positive = []
negative = []

for i in priceChange:
    if i > 0:
        positive.append(i)
        print(positive) # show what indentation does
```

To indicate a block of code in Python, you must indent each line of the block by the same amount.

```python
positive = []
negative = []

for i in priceChange:
    if i > 0:
        positive.append(i)
    else:
        negative.append(i)
        
print(positive, negative)
```

```python
mylist = []

for i in range(1,10): # show what happens after changing 1 too 0
    mylist.append(i)
    
mylist
```
